```python
import cv2
from google.colab.patches import cv2_imshow
```

```python
# ready model trained to detect the face in the picture
# does processing on the objects in the picture (object detection)
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
```

```python

image = cv2.imread('image.jpg')
```

```python
# convert the photo to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
# detectMultiScale -> fn use ML algorithm to scan the picture to detect the face
# gray -> we reduced the dimensions
# scaleFactor -> scale the picture
# minNeighbors -> dont detect that this a face unless there is 5 neighbor nodes say that this is a face, when we inc minNeighbors we inc the accuracy
# minSize -> min size of the picture
faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(40, 40))
```

```python
# the rectangle that detect the face
# x, y -> coordinates of the rectangle
# (255, 0, 0) -> color
# 2 -> thickness
for (x, y, w, h) in faces:
    cv2.rectangle(image, (x, y), (x+w, y+h), (255, 0, 0) , 2)
```
